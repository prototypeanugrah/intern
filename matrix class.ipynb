```python
import numpy as np
class matrix(object):
    
    def __init__(self,mat):
        self.mat = mat
        self.row = len(mat)
        self.col = len(mat[0])
        self.resultant = []
    
    def __add__(self, other):
        #input: matrix(self) and matrix(other)
        #output: (matrix)
        self.resultant = [[0 for c in range(self.col)] for r in range(self.row)]
        for i in range(self.row):
            for j in range(other.col):
                self.resultant[i][j] = self.mat[i][j] + other.mat[i][j]
        return matrix(self.resultant)

    def __sub__(self, other):
        #input: matrix(self) and matrix(other)
        #output: (matrix)
        self.resultant = [[0 for r in range(self.col)] for c in range(self.row)]
        for i in range(self.row):
            for j in range(self.col):
                self.resultant[i][j] = self.mat[i][j] - other.mat[i][j]
        return matrix(self.resultant)
    
    def __mul__(self, other):
        #input: matrix(self) and matrix(other)
        #output: (matrix)
        self.resultant = [[0 for c in range(other.col)] for r in range(self.row)]
        for i in range(self.row):
            for j in range(other.col):
                for k in range(other.row):
                    self.resultant[i][j] += self.mat[i][k] * other.mat[k][j]
        return matrix(self.resultant)
    
    def __repr__(self):
        return f'{self.mat}'
    
    def determinant(self):
        #input: matrix(self) 
        #output: (float)
        print("determinant : " + str(np.linalg.det(self.mat)))
        
    def l2_norm(self):
        #input: matrix(self)
        #ouput: (float)
        #print(type(self))
        print("l2 norm : " + str(np.linalg.norm(self.mat)))
    
    def matrix_vector_multiplication(self, other):
        #input: matrix(self) and list(other)
        #output: (matrix)
        return matrix(np.dot(self.mat, other))

    def compare_matrices(self, other):
        #input: matrix(self) and matrix(other)
        #output: (boolean)
        for i in range(self.row):
            for j in range(self.col):
                if self.mat[i][j] == other.mat[i][j]:
                    return True
                else:
                    return False
```
