```python
import math
class vector:
    
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
    
    def __add__(self,other):
        #input: self (vector), other(vector)
        #output: (vector)
        return vector(self.x+other.x,self.y+other.y,self.z+other.z)
    
    def __sub__(self,other):
        #input: self (vector), other(vector)
        #output: (vector)
        return vector(self.x-other.x,self.y-other.y,self.z-other.z)
    
    def determinant(self):
        #input: self (vector)
        #output: (float)
        return math.sqrt(pow(self.x,2)+pow(self.y,2)+pow(self.z,2))
    
    def __mul__(self,other):
        #input: self (vector), other(vector)
        #output: (vector)
        return vector(self.x*other.x,self.y*other.y,self.z*other.z)
    
    def __repr__(self):
        #input: self (vector)
        #output: (string)
        return f'({self.x},{self.y},{self.z})'
        #return '(' + str(self.x) + ',' + str(self.y) + ',' + str(self.z) + ')'
    
    def scalar_multiplication(self,r):
        #input: self (vector), other(integer)
        #output: (vector)
        return vector(r*self.x,r*self.y,r*self.z)
    
    def compare_equal_vectors(self,other):
        #input: self(vector) and other(vector)
        #output: (boolean)
        return self.x==other.x and self.y == other.y and self.z == other.z 
    
    def compare_greater_vector(self,other):
        #input: self(vector) and other(vector)
        #output: (boolean)
        return vector.determinant(self)>vector.determinant(other)
    
    def compare_smaller_vector(self,other):
        #input:self(vector) and other(vector)
        #output:(boolean)
        return vector.determinant(self)<vector.determinant(other)
```

```python
def test_addition(): 
    #input: vector (1,2) and vector(3,4)
    #output: vector(4,6)
    v1 = vector(1,2,3)
    v2 = vector(3,4,5)
    v3 = vector(4,6,8)
    v4 = v1 + v2 
    print(v3.compare_equal_vectors(v4))

def test_subtraction():
    #input: vector(3,6,2) and vector(4,2,5)
    #output: vector(-1,-4,3)
    v1 = vector(3,6,2)
    v2 = vector(4,2,5)
    v3 = vector(-1,-4,3)
    v4 = v1 - v2
    print(v4)
    print(v3.compare_equal_vectors(v4))
    
def test_scalar_multiplication():
    #input: vector(2,4,3) and scalar(2)
    #output: vector(4,8,6)
    v1 = vector(2,4,3)
    a = int(input("enter an integer"))
    print(v1.scalar_multiplication(a))

def test_element_wise_multiplication():
    #input: vector(4,3,5) and vector(3,5,4)
    #output: vector(12,15,20)
    v1 = vector(4,3,5)
    v2 = vector(3,5,4)
    v3 = vector(12,15,20)
    v4 = v1 * v2
    print(v3.compare_equal_vectors(v4))

test_addition()
test_subtraction()
test_scalar_multiplication()
test_element_wise_multiplication()
```
